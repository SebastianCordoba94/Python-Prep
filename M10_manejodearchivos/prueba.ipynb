```python
import datetime

x = datetime.datetime.now()
print("Ahora =",x)
hora = str(x)
#hora = datetime.strptime(x, '%Y-%m-%d %H:%M:%S')
print("Ahora =",hora)

fecha_hora = x.day,"-",x.month,"-",x.year,x.hour,":",x.minute,":",x.second
print("Fecha hora ",fecha_hora)
objeto_datetime = datetime.datetime.strptime(hora, '%Y-%m-%d %H:%M:%S')
print("objeto datetime =", objeto_datetime)
marca_de_tiempo = datetime.datetime.timestamp(objeto_datetime)
print("timestamp =", marca_de_tiempo)
fecha_hora2 = datetime.datetime.fromtimestamp(marca_de_tiempo)
print("fecha hora =", fecha_hora2)
```
